```python
# lets first import the  required library for our exploratory data analysis.
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
#Importing the dataset
data = pd.read_csv('penguins_size.csv')
```

```python
# We first use head method of pandas to take a look of data.
data.head()
```

```python
# We then use pandas info method to check the  number of null values in the  dataset.It also helps us in finding datatypes of different attributes
 
data.info()
```

```python
# pandas method shape is  used to check the number rows and columns.
data.shape
```

```python
# We use the dropna method of pandas to remove missing values and  store the dataframe return by the method in 
data2 = data.dropna()
```

```python
# Now you can see our new dataframe
data2.info()
```

```python
# We will use seaborn's countplot to see the number of different species we have avalible in our data.
# you can see the plot shows we have twice the number of observations related to the 'Adelie' specie than 'Chinstrap'
sb.set(rc = {'figure.figsize':(15,8)})
sb.countplot(x='species',data=data2)
```

```python
# Now we will use seaborn's scatterplot to see if penguins culmen length and depth is enough to seperate them.
# As you can see the culmen_length and culmen_depth is a good way to differentiate penguin species. 
sb.scatterplot(data=data2,x='culmen_length_mm',y='culmen_depth_mm',hue='species')
```

```python
# We use unique method to find out about the different island in attribute 'island'.
data2['island'].unique()
```

```python
#We will again use seaborn's countplot to find about how many  observations we have about different penguin populated islands. 
sb.countplot(x='island',data=data2)
```

```python
# we will use scatterpot to compare how bodymass and  culmens_length can be used to differentiate penguin species.
sb.scatterplot(data=data2,x='body_mass_g',y='culmen_length_mm',hue='species') 
```
